```python
import pandas as pd

# 1. Загрузка данных из CSV
file_path = r"C:\Users\elena.mordvinova\PYT\L1\1_Airline Delay-задержка рейса .csv"
data = pd.read_csv(file_path)

# Преобразование 'FlightDate' в формат даты
data['FlightDate'] = pd.to_datetime(data['FlightDate'])

# 1. Применение функции через apply()
def delay_summary(group):
    return pd.Series({
        'mean_arrival_delay': group['ArrivalDelay'].mean(),
        'mean_departure_delay': group['DepartureDelay'].mean()
    })

grouped_data_apply = data.groupby('FlightDate').apply(delay_summary)
print(grouped_data_apply)

# 2. Применение функции через transform()
def mean_delay(group):
    return group.mean()

grouped_data_transform = data.groupby('FlightDate')[['ArrivalDelay', 'DepartureDelay']].transform(mean_delay)
data[['Mean_Arrival_Delay', 'Mean_Departure_Delay']] = grouped_data_transform

print(data[['FlightDate', 'ArrivalDelay', 'Mean_Arrival_Delay', 'DepartureDelay', 'Mean_Departure_Delay']].head())

# 3. Применение функции через map()
def adjust_delay(delay):
    return delay + 10  # Добавление 10 минут задержки

data['Adjusted_Arrival_Delay'] = data['ArrivalDelay'].map(adjust_delay)
data['Adjusted_Departure_Delay'] = data['DepartureDelay'].map(adjust_delay)

print(data[['FlightDate', 'ArrivalDelay', 'Adjusted_Arrival_Delay']].head())

# 4. Ресемплирование данных по дате
data.set_index('FlightDate', inplace=True)  # Установка индекса
resampled_data = data.resample('M').agg({
    'ArrivalDelay': 'mean',
    'DepartureDelay': 'mean'
})

print(resampled_data)
```

```python
import pandas as pd

# Загрузка данных из CSV файла
file_path = r"C:\Users\elena.mordvinova\PYT\L1\1_Airline Delay-задержка рейса .csv"
data = pd.read_csv(file_path)

# Проверка названий колонок
print("Названия колонок:", data.columns)

# Удаление лишних пробелов из названий колонок
data.columns = data.columns.str.strip()

# Создание колонки 'FlightDate' на основе 'year' и 'month'
data['FlightDate'] = pd.to_datetime(data[['year', 'month']].assign(day=1))

# 1. Применение функции через apply()
def delay_summary(group):
    return pd.Series({
        'mean_arr_delay': group['arr_delay'].mean(),
        'total_flights': group['arr_flights'].sum()
    })

grouped_data_apply = data.groupby('FlightDate').apply(delay_summary)
print("Результаты apply:\n", grouped_data_apply)

# 2. Применение функции через transform()
def mean_delay(group):
    return group.mean()

grouped_data_transform = data.groupby('FlightDate')[['arr_delay', 'arr_flights']].transform(mean_delay)
data[['Mean_Arr_Delay', 'Total_Arr_Flights']] = grouped_data_transform

print("Данные с добавленными средними значениями:\n", data[['FlightDate', 'arr_delay', 'Mean_Arr_Delay', 'arr_flights', 'Total_Arr_Flights']].head())

# 3. Применение функции через map()
def adjust_delay(delay):
    return delay + 10  # Добавление 10 минут к задержкам

data['Adjusted_Arr_Delay'] = data['arr_delay'].map(adjust_delay)

print("Данные с отредактированными задержками:\n", data[['FlightDate', 'arr_delay', 'Adjusted_Arr_Delay']].head())

# 4. Ресемплирование данных по дате
data.set_index('FlightDate', inplace=True)  # Установка индекса
resampled_data = data.resample('M').agg({
    'arr_delay': 'mean',
    'arr_flights': 'sum'
})

print("Ресемплированные данные по месяцам:\n", resampled_data)

# 5. Подключение к БД dvdrental (если нужно)
# import sqlalchemy
# engine = sqlalchemy.create_engine('postgresql://user:password@localhost/dvdrental')
# data_dvdrental = pd.read_sql('SELECT * FROM rental', engine.connect())
# connection.close()
```
