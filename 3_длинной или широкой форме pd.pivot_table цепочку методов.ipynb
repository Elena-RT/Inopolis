```python
import pandas as pd

# Ответ: На шаге (3) мы преобразовали данные в длинный формат для удобства анализа и визуализации задержек по авиакомпаниям. На шаге (4) рассчитали средние задержки, чтобы выявить проблемы и улучшить качество обслуживания операторов.

# 1. Загрузка данных из CSV файла
file_path = r"C:\Users\elena.mordvinova\PYT\L1\1_Airline Delay-задержка рейса .csv"
data = pd.read_csv(file_path)  # Чтение данных из файла

# 2. Изучение структуры данных
print(data.head())  # Вывод первых 5 строк
print(data.info())  # Информация о типах данных и количестве строк

# 3. Преобразование данных в длинный формат (если требуется)
data_long = data.melt(id_vars=['Column1', 'Column2'], var_name='Airline', value_name='Delay')  
# Преобразование с указанием столбцов, которые остаются фиксированными

# 4. Создание сводной таблицы для средней задержки по авиакомпаниям
pivot_table = pd.pivot_table(data_long, values='Delay', index='Airline', aggfunc='mean')  
print("Средняя задержка по авиакомпаниям:\n", pivot_table)

# 5. Группировка данных по авиакомпаниям и вычисление средних задержек
grouped_data = data_long.groupby('Airline').agg(
    Average_Arrival_Delay=('ArrivalDelay', 'mean'),  # Средняя задержка прибытия
    Average_Departure_Delay=('DepartureDelay', 'mean')  # Средняя задержка вылета
).reset_index()  

print("Средние задержки по прибытиям и вылетам:\n", grouped_data)
```
