```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from sklearn.linear_model import LinearRegression

# 1. Загружаем данные из CSV файла
file_path = r"C:\Users\elena.mordvinova\PYT\L1\1_Airline Delay-задержка рейса .csv"
data = pd.read_csv(file_path)

# Просмотр первых строк данных для понимания структуры
print(data.head())

# Очистка данных от выбросов и пропусков
data_clean = data.dropna()  # Удаляем строки с пропусками
data_clean = data_clean[(np.abs(stats.zscore(data_clean['колонка_с_данными'])) < 3)]  # Замените на реальное название столбца

# 2. Выбор количественных переменных
# Замените 'переменная1' и 'переменная2' на реальные названия столбцов вашего очищенного датафрейма
var1 = data_clean['переменная1']
var2 = data_clean['переменная2']

# 5. Построение точечной диаграммы и оценка корреляции
plt.figure(figsize=(10, 6))
plt.scatter(var1, var2, alpha=0.6)
plt.xlabel('Переменная 1')
plt.ylabel('Переменная 2')
plt.title('Точечная диаграмма переменных 1 и 2')
plt.grid(True)
plt.show()

# Оценка корреляции
correlation = stats.pearsonr(var1, var2)
print(f'Коэффициент корреляции Пирсона: {correlation[0]:.3f}, p-значение: {correlation[1]:.3f}')
if correlation[1] < 0.05:
    print("Корреляция статистически значима.")
else:
    print("Корреляция не является статистически значимой.")

# 6. Оценка модели линейной регрессии

# Преобразуем данные для регрессии
X = var1.values.reshape(-1, 1)  # Преобразуем в двумерный массив для модели
y = var2.values

# Создаем и обучаем модель линейной регрессии
model = LinearRegression().fit(X, y)
predictions = model.predict(X)

# Визуализируем результаты регрессии
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', alpha=0.6, label='Данные')
plt.plot(X, predictions, color='red', linewidth=2, label='Линейная регрессия')
plt.title('Линейная регрессия между переменной 1 и переменной 2')
plt.xlabel('Переменная 1')
plt.ylabel('Переменная 2')
plt.legend()
plt.grid(True)
plt.show()

# Вывод коэффициентов и смещения модели
print(f'Коэффициенты модели: {model.coef_[0]:.3f}, Смещение: {model.intercept_:.3f}')
#вывод:регрессия помогает прогнозировать продажи на основе ключевых факторов, что позволяет оптимизировать продажи\стратегии и принимать обоснованные решения для увеличения прибыли
```
